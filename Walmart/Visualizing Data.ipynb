```python
# Import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import ggplot, aes, geom_bar, scale_x_discrete, theme
```

```python
# Load dataset
walmart_data = pd.read_csv("data/Walmart.csv")
walmart_data.head()
```

```python
walmart_data_sorted = walmart_data.sort_values('Weekly_Sales', ascending = False)
walmart_data_sorted
```

### Plot a bar chart using Matplotlib

```python
plt.figure(figsize=(12, 6))

x = walmart_data_sorted['Store'][0:10]
y = walmart_data_sorted['Weekly_Sales'][0:10]

plt.bar(x, y)
plt.xlabel('Store', fontsize=12)
plt.ylabel('Weekly Sales', fontsize=12)
plt.title('Top 10 Stores with the Highest Weekly Sales', fontsize=15)

# Set specific x-axis tick positions using set_xticks
plt.xticks(walmart_data_sorted['Store'][0:10], fontsize=10)

plt.tight_layout()

plt.show()
```

### Create Subplots

```python
# Get data for the top 10 stores
top_stores_data = walmart_data_sorted.head(10)

fig, ax = plt.subplots(figsize=(40, 18), ncols=2)

# First subplot
ax[0].bar(top_stores_data['Store'], top_stores_data['Weekly_Sales'])
ax[0].set_xticks(top_stores_data['Store'])
ax[0].set_xlabel('Store', fontsize=25)
ax[0].set_ylabel('Weekly sales in millions', fontsize=25)
ax[0].set_title('Top 10 Stores with the Highest Weekly Sales', fontsize=25)
ax[0].tick_params(axis='both', labelsize=20)

# Second subplot
ax[1].bar(top_stores_data['Store'], top_stores_data['Temperature'])
ax[1].set_xticks(top_stores_data['Store'])
ax[1].set_xlabel('Store', fontsize=25)
ax[1].set_ylabel('Temperature', fontsize=25)
ax[1].set_title('Top 10 Stores with the Highest Weekly Sales per Temperature', fontsize=25)
ax[1].tick_params(axis='both', labelsize=20)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))

data = walmart_data_sorted.head(10)  # Get the top 10 stores

sns.barplot(data=data, x='Store', y='Weekly_Sales')  # Use the correct column name 'Weekly_Sales'

plt.xlabel('Store', fontsize=13)
plt.ylabel('Weekly Sales in millions', fontsize=13)
plt.title('Top 10 Stores with the Highest Weekly Sales', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.tight_layout()

plt.show()
```
