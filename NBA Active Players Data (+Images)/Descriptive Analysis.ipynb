```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load the dataset
nba = pd.read_csv('data/players.csv')
```

```python
# Calculate the average height and weight
average_height = nba['height'].mean()
average_weight = nba['weight'].mean()

print(f"Average height of players: {average_height:.2f} inches")
print(f"Average weight of players: {average_weight:.2f} lbs")
```

```python
# Create a box plot for height by position
plt.figure(figsize=(10, 6))
sns.boxplot(data=nba, x='position', y='height')
plt.title('Height Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Height (inches)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert the birthday column to datetime dtype
nba['birthday'] = pd.to_datetime(nba['birthday'])

# Calculate age based on the 'birthday' column
current_year = datetime.now().year
nba['age'] = current_year - nba['birthday'].dt.year

# Create a histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=nba, x='age', bins=20, kde=True)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()

# Calculate the average age of players
average_age = nba['age'].mean()
print(f"Average age of players: {average_age:.2f} years")
```

```python
# Find the most common draft round
most_common_draft_round = nba['draft_round'].mode()[0]

# Find the most common draft number
most_common_draft_number = nba['draft_number'].mode()[0]

print(f"The most common draft round: {most_common_draft_round}")
print(f"The most common draft number: {most_common_draft_number}")
```
