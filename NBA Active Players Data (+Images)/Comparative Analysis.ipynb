```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Load the dataset
nba = pd.read_csv('data/players.csv')
```

```python
# Create a scatter plot for height vs. weight
plt.figure(figsize=(10, 6))
plt.scatter(nba['height'], nba['weight'], alpha=0.5)
plt.title('Correlation between Height and Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.show()
```

```python
# Calculate the average height and weight for each country
country_stats = nba.groupby('country').agg({'height': 'mean', 'weight': 'mean'}).reset_index()
country_stats
```

```python
# Get the top 10 countries
top_countries = nba['country'].value_counts().head(10).index

# Filter data for top 10 countries
top_countries_data = nba[nba['country'].isin(top_countries)]

# Create a countplot for player positions by country
plt.figure(figsize=(12, 8))
sns.countplot(data=top_countries_data, x='country', hue='position')
plt.title('Distribution of Players by Position in Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.legend(title='Position')
plt.show()
```

```python
# Calculate the average height of players drafted in different years
average_height_by_year = nba.groupby('draft_year')['height'].mean()

# Create a line plot to visualize the trend
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_height_by_year.index, y=average_height_by_year.values)
plt.title('Average Height of Players Drafted by Year')
plt.xlabel('Draft Year')
plt.ylabel('Average Height (inches)')
plt.xticks(rotation=45)
plt.show()
```
