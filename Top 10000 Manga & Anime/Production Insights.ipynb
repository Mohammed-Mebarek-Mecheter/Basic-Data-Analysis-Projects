```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Dataset
anime_df = pd.read_csv("data/clean_anime.csv")
manga_df = pd.read_csv("data/clean_manga.csv")
```

```python
# Count the occurrences of each studio
studio_counts = anime_df['Studios'].value_counts()

# Select the top 10 studios
top_10_studios = studio_counts.head(10)

# Plot the studio distribution
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_studios.values, y=top_10_studios.index, palette='viridis')
plt.title('Top 10 Prolific Animation Studios')
plt.xlabel('Count')
plt.ylabel('Studio')
plt.show()
```

```python
# Clean up the 'Producers' column (including removing empty strings)
anime_df['Producers'] = anime_df['Producers'].apply(lambda x: [producer.strip(" '[]") for producer in x.split(', ') if producer.strip()])

# Exclude rows with empty lists or containing only empty strings
anime_df = anime_df[anime_df['Producers'].apply(lambda x: len(x) > 0 and all(producer != '' and producer != 'add some' for producer in x))]

# Count the occurrences of each production company
producer_counts = anime_df['Producers'].explode().value_counts()

# Select the top 10 production companies
top_10_producers = producer_counts.head(10)

# Plot the production company distribution
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_producers.values, y=top_10_producers.index, palette='viridis')
plt.title('Top 10 Prolific Production Companies')
plt.xlabel('Count')
plt.ylabel('Production Company')
plt.show()
```

```python
# Count the occurrences of each serialization platform
serialization_counts = manga_df['Serialization'].value_counts()

# Select the top 10 serialization platforms
top_10_serialization = serialization_counts.head(10)

# Plot the serialization impact
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_serialization.values, y=top_10_serialization.index, palette='viridis')
plt.title('Top 10 Manga Serialization Platforms and Their Impact')
plt.xlabel('Count')
plt.ylabel('Serialization Platform')
plt.show()
```
