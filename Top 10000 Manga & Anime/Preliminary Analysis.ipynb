```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import Dataset
anime_df = pd.read_csv("data/clean_anime.csv")
manga_df = pd.read_csv("data/clean_manga.csv")
```

```python
# Identify the top-rated anime based on user scores
print("Top 10 Rated Anime:")
top_anime = anime_df.sort_values(by="Score", ascending=False).head(10)
top_anime[['Title', 'Score']]
```

```python
# Identify the top-rated manga based on user scores
print("Top 10 Rated Manga:")
top_manga = manga_df.sort_values(by="Score", ascending=False).head(10)
top_manga[['Title', 'Score']]
```

##### Relationship between User Scores and Number of Votes

```python
# Scatter plot with seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Vote', y='Score', data=anime_df, alpha=0.7)
plt.title('Relationship between User Scores and Number of Votes for Anime')
plt.xlabel('Number of Votes')
plt.ylabel('User Score')
plt.show()
```

```python
# Explore the relationship between user scores and the number of votes for manga
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Vote', y='Score', data=manga_df, alpha=0.7)
plt.title('Relationship between User Scores and Number of Votes for Manga')
plt.xlabel('Number of Votes')
plt.ylabel('User Score')
plt.show()
```

##### Influence of Source Material on Anime Scores

```python
# Calculate average scores for each source material
average_scores_by_source = anime_df.groupby('Source')['Score'].mean().sort_values(ascending=False)

# Print the results
print("Average Scores by Source Material:")
average_scores_by_source
```

```python
# Box plot for anime source material influence
plt.figure(figsize=(12, 8))
sns.boxplot(x='Source', y='Score', data=anime_df)
plt.title('Influence of Source Material on Anime Scores')
plt.xlabel('Source Material')
plt.ylabel('User Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Bar plot for anime source material influence
plt.figure(figsize=(12, 8))
sns.barplot(x=average_scores_by_source.index, y=average_scores_by_source.values)
plt.title('Influence of Source Material on Anime Scores')
plt.xlabel('Source Material')
plt.ylabel('Average User Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```
