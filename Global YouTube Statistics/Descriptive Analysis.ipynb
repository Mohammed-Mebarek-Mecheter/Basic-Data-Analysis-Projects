```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("data/Global YouTube Statistics.csv", encoding='latin1')
```

##### Subscriber distribution

```python
# Create a histogram of subscriber counts
plt.figure(figsize=(10, 6))
plt.hist(data['subscribers'], bins=20, color='skyblue', edgecolor='black')
plt.title('Subscriber Distribution')
plt.xlabel('Number of Subscribers')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Set the style of Seaborn
sns.set(style="whitegrid")

# Create a histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data['subscribers'], bins=20, kde=True, color='skyblue')
plt.title('Subscriber Distribution')
plt.xlabel('Number of Subscribers')
plt.ylabel('Frequency')
plt.show()
```

#### Video views distribution

```python
# Create a histogram of video views
plt.figure(figsize=(10, 6))
plt.hist(data['video views'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Video Views Distribution')
plt.xlabel('Number of Video Views')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Set the style of Seaborn
sns.set(style="whitegrid")

# Create a histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data['video views'], bins=20, kde=True, color='lightgreen')
plt.title('Video Views Distribution')
plt.xlabel('Number of Video Views')
plt.ylabel('Frequency')
plt.show()
```

##### Category distribution

```python
# Set the style of Seaborn
sns.set(style="whitegrid")

# Create a countplot for category distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='category', order=data['category'].value_counts().index, palette='viridis')
plt.title('Category Distribution')
plt.xlabel('Number of Channels')
plt.ylabel('Category')
plt.show()
```

##### Upload frequency analysis

```python
# Set the style of Seaborn
sns.set(style="whitegrid")

# Create a histogram of upload frequency
plt.figure(figsize=(10, 6))
sns.histplot(data['uploads'], bins=20, kde=True, color='salmon')
plt.title('Upload Frequency Analysis')
plt.xlabel('Number of Uploads')
plt.ylabel('Frequency')
plt.show()
```

##### Country-wise distribution

```python
# Set the style of Seaborn
sns.set(style="whitegrid")

# Get the top 10 countries by channel count
top_countries = data['Country'].value_counts().head(10)

# Create a bar plot for country-wise distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='coolwarm')
plt.title('Top 10 Country-wise Distribution')
plt.xlabel('Country')
plt.ylabel('Number of Channels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sort the data by subscribers in descending order
sorted_data = data.sort_values(by='subscribers', ascending=False)

# Select the top 100 most subscribed channels
top_subscribed = sorted_data.head(100)

# Group by category and calculate the average subscribers
category_avg_subscribers = top_subscribed.groupby('category')['subscribers'].mean().sort_values(ascending=False)
category_avg_subscribers
```

```python
# Create a bar plot for top categories/niches
plt.figure(figsize=(12, 6))
sns.barplot(x=category_avg_subscribers.index, y=category_avg_subscribers.values, palette='viridis')
plt.title('Top Categories Among Most Subscribed YouTube Channels')
plt.xlabel('Category')
plt.ylabel('Average Subscribers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a box plot for subscribers distribution by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='subscribers', data=data, palette='Set3')
plt.title('Distribution of Subscribers Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a violin plot for video views distribution by category
plt.figure(figsize=(12, 6))
sns.violinplot(x='category', y='video views', data=data, palette='Set3')
plt.title('Distribution of Video Views Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Number of Video Views')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by category and calculate the average upload frequency
category_avg_uploads = data.groupby('category')['uploads'].mean().sort_values(ascending=False)
category_avg_uploads
```

```python
# Create a bar plot for average upload frequency by category
plt.figure(figsize=(12, 6))
sns.barplot(x=category_avg_uploads.index, y=category_avg_uploads.values, palette='Set3')
plt.title('Average Upload Frequency of YouTube Channels in Different Categories')
plt.xlabel('Category')
plt.ylabel('Average Upload Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Get the top 20 countries by channel count
top_countries = data['Country'].value_counts().head(20).index
top_countries
```

```python
# Create a box plot for subscribers distribution by top 20 countries
plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='subscribers', data=data[data['Country'].isin(top_countries)], palette='Set3')
plt.title('Distribution of Subscribers by Top 20 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
