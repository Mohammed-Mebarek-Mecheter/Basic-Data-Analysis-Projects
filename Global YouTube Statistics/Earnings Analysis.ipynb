```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("data/Global YouTube Statistics.csv", encoding='latin1')
```

```python
# Calculate the average earnings for each channel type
earnings_by_channel_type = data.groupby('channel_type')[['lowest_yearly_earnings', 'highest_yearly_earnings']].mean()
earnings_by_channel_type
```

```python
# Group by channel type and calculate the mean earnings
earnings_by_type = data.groupby('channel_type')[['lowest_yearly_earnings', 'highest_yearly_earnings']].mean()
earnings_by_type
```

```python
# Create a bar plot for earnings by channel type
plt.figure(figsize=(10, 6))
earnings_by_type.plot(kind='bar', colormap='coolwarm', rot=0)
plt.title('Earnings Variation by Channel Type')
plt.xlabel('Channel Type')
plt.ylabel('Average Yearly Earnings')
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean earnings for top 5 channel types
top_channel_types = data['channel_type'].value_counts().head(5).index
earnings_by_channel_type = data[data['channel_type'].isin(top_channel_types)].groupby('channel_type')['highest_yearly_earnings'].mean()
earnings_by_channel_type
```

```python
# Create a bar plot for earnings by channel type
plt.figure(figsize=(10, 6))
sns.barplot(x=earnings_by_channel_type.index, y=earnings_by_channel_type.values, palette='pastel')
plt.title('Earnings Variation by Top 5 Channel Types')
plt.xlabel('Channel Type')
plt.ylabel('Mean Highest Yearly Earnings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by 'created_date' and calculate the mean earnings
earnings_over_time = data.groupby('created_date')['highest_yearly_earnings'].mean()
earnings_over_time
```

```python
# Create a line plot for earnings over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=earnings_over_time, color='blue')
plt.title('Earnings Over Time')
plt.xlabel('Date')
plt.ylabel('Mean Highest Yearly Earnings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate the median and average monthly earnings
median_monthly_earnings = data['lowest_monthly_earnings'].median()
average_monthly_earnings = data['lowest_monthly_earnings'].mean()

# Calculate the median and average yearly earnings
median_yearly_earnings = data['lowest_yearly_earnings'].median()
average_yearly_earnings = data['lowest_yearly_earnings'].mean()

print(f"Median Monthly Earnings: {median_monthly_earnings:.2f}")
print(f"Average Monthly Earnings: {average_monthly_earnings:.2f}")
print(f"Median Yearly Earnings: {median_yearly_earnings:.2f}")
print(f"Average Yearly Earnings: {average_yearly_earnings:.2f}")
```

```python
# Calculate the difference between highest and lowest monthly earnings
monthly_earnings_difference = data['highest_monthly_earnings'] - data['lowest_monthly_earnings']

# Calculate the difference between highest and lowest yearly earnings
yearly_earnings_difference = data['highest_yearly_earnings'] - data['lowest_yearly_earnings']

# Calculate the average differences
average_monthly_earnings_difference = monthly_earnings_difference.mean()
average_yearly_earnings_difference = yearly_earnings_difference.mean()

print(f"Average Difference in Monthly Earnings: {average_monthly_earnings_difference:.2f}")
print(f"Average Difference in Yearly Earnings: {average_yearly_earnings_difference:.2f}")
```
