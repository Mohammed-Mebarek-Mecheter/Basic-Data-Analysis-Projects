```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("data/Global YouTube Statistics.csv", encoding='latin1')
```

```python
# Filter data for the last 15 years
current_year = 2023
last_15_years = current_year - 15
filtered_data = data[data['created_year'] >= last_15_years]

# Group data by created_year and calculate metrics
trends_over_time = filtered_data.groupby('created_year').agg({
    'subscribers_for_last_30_days': 'sum',
    'video views': 'sum',
    'highest_yearly_earnings': 'sum'
}).reset_index()

# Create line plots for trends over time
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.lineplot(x='created_year', y='subscribers_for_last_30_days', data=trends_over_time)
plt.title('Subscribers Gained Over Time')
plt.xlabel('Year')
plt.ylabel('Subscribers Gained')

plt.subplot(1, 3, 2)
sns.lineplot(x='created_year', y='video views', data=trends_over_time)
plt.title('Video Views Over Time')
plt.xlabel('Year')
plt.ylabel('Video Views')

plt.subplot(1, 3, 3)
sns.lineplot(x='created_year', y='highest_yearly_earnings', data=trends_over_time)
plt.title('Earnings Over Time')
plt.xlabel('Year')
plt.ylabel('Earnings')

plt.tight_layout()
plt.show()
```
