```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read the CSV file
df = pd.read_csv("data/netflix_clean.csv")
```

```python
df.head()
```

##### Distribution of Movies and TV Shows

```python
# Count the number of movies and TV shows
show_counts = df['type'].value_counts()

# Plot the distribution of movies and TV shows
plt.figure(figsize=(8, 6))
show_counts.plot(kind='bar', color=['blue', 'orange'])
plt.title('Distribution of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

##### Distribution of release years

```python
# Plot the distribution of release years
plt.figure(figsize=(12, 6))
plt.hist(df['release_year'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

##### Temporal trend of content availability

```python
# Group the data by release year and count the number of movies/TV shows released each year
release_year_counts = df['release_year'].value_counts().sort_index()

# Plot the temporal trend of content availability
plt.figure(figsize=(12, 6))
plt.plot(release_year_counts.index, release_year_counts.values, marker='o', color='orange')
plt.title('Temporal Trend of Content Availability')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies/TV Shows Released')
plt.grid(True)
plt.show()
```

##### Distribution of content across different countries

```python
# Filter out rows with 'Unknown' values
df_filtered = df[df['country'] != 'Unknown']
# Count the number of entries for each country
country_counts = df_filtered['country'].value_counts().head(10)  # Consider only the top 10 producing countries

# Plot the distribution of content across different countries
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Producing Countries of Netflix Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter out rows with 'Unknown' values
df_filtered = df[df['rating'] != 'Unknown']
# Count the number of entries for each content rating
rating_counts = df_filtered['rating'].value_counts()

# Plot the distribution of content ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=rating_counts.index)
plt.title('Distribution of Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
